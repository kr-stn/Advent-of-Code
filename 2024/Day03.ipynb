# Day 3
- multiply tuples of 1-3 digit INTs
- allowed input: "mul(X,Y)"
- everything else is not allowed
- string parsing and finding allowed instances (maybe with regex)

```python
# read lines, then split into list of lists by space separator
with open("data/day03.txt", "r") as in_file:
    in_string = in_file.read().strip()
```

```python
import math
import re
```

```python
mults = re.findall(r'mul\(\d{1,3},\d{1,3}\)', in_string)
print(mults)
```

```python
mult_dict = [[int(x) for x in re.findall(r'\d{1,3}',y)] for y in mults]
```

```python
total_mult = sum(math.prod(x) for x in mult_dict)
print("Sum of all multiplies:", total_mult)
```

## Part 2
- only use the multiply strings that are between "do" and "don't" statements
- can filter the string to only those sections and then use the same logic as before

```python
# add do and don't at start and end
# match strings between do and don't() and then filter with same logic as before
valid = "".join(re.findall(r"do\(\)(.*?)don't\(\)", "do()" + in_string + "don't()"))

mults = re.findall(r'mul\(\d{1,3},\d{1,3}\)', valid)
mult_dict = [[int(x) for x in re.findall(r'\d{1,3}',y)] for y in mults]

total_mult = sum(math.prod(x) for x in mult_dict)
print("Sum of all valid multiplies:", total_mult)
```
